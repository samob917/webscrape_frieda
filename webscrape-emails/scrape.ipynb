```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import time
```

```python
from selenium.webdriver.common.by import By
from selenium import webdriver
from selenium.webdriver.chrome.options import Options

# Create an Options object
chrome_options = Options()

# Add the headless argument
chrome_options.add_argument("--headless=new") # For newer Chrome versions
# For older Chrome versions, use: chrome_options.add_argument("--headless")
driver = webdriver.Chrome(options=chrome_options)
```

```python
def make_url(page, loc, spec):
    return f"https://freida.ama-assn.org/search/list?spec={spec}&loc={loc}&page={page}"

def get_search_results(url):
    searchResults = []
    driver.get(url)
    searches = driver.find_elements(By.CLASS_NAME, "ama-card")
    for search in searches:
        searchResults.append(search.text)
    return searchResults

def get_details(URL):
    details = {}
    page = requests.get(URL)
    soup = BeautifulSoup(page.content, "html.parser")
    details["title"] = soup.find("title").text.split("|")[0].strip()
    pText = page.text
    contactText = pText[pText.find('"parent_field_name":"field_program_director"'):]
    try:
        details["first_name"] = contactText[contactText.find("field_first_name"):].split('field_first_name":"')[1].split('"')[0]
    except:
        details["first_name"] = "UNKNOWN"
    try: 
        details["last_name"] = contactText[contactText.find("field_last_name"):].split('field_last_name":"')[1].split('"')[0]
    except:
        details["last_name"] = "UNKNOWN"
    try:
        details["email"] = contactText[contactText.find("field_email"):].split('field_email":"')[1].split('"')[0]
    except:
        details["email"] = "UNKNOWN"
    try:
        details["org"] = contactText[contactText.find("organization"):].split('organization":"')[1].split('"')[0]
    except:
        details["org"] = "UNKNOWN"
    return details
```

```python
def get_csv_for_spec(spec_code):
    loc=["01","02","03","04","05","06","07","08","09","PR"]
    spec=spec_code
    result_list = []
    for l in loc:
        page=1
        url = make_url(page, l, spec)
        print(url)
        searchResults = get_search_results(url)
        print(searchResults)
        
        while len(searchResults) > 0:
            for search in searchResults:
                print(search)
                time.sleep(1)
                splitSearch = search.split("\n")
                print(splitSearch)
                title = splitSearch[1]
                specialty = splitSearch[0]
                id = splitSearch[3].split(": ")[1]
                detail_dict = get_details(f"https://freida.ama-assn.org/program/{id}")
                detail_dict["title"] = title
                detail_dict["specialty"] = specialty
                result_list.append(detail_dict)
            page += 1
            searchResults = get_search_results(make_url(page, loc, spec))
    spec_df = pd.DataFrame(result_list)
    print("Num records:", len(spec_df))
    spec_df.to_csv(f"{spec_code}_residency_programs.csv")
```
