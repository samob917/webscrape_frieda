```python
def get_random_details(df):
    return df.iloc[int(np.floor(len(df) * np.random.random()))]
```

```python
def get_random_template(templates):
    template = templates[int(np.floor(len(templates) * np.random.random()))]
    return template
```

```python
def send_n_emails(sent_csv, new_list, templates, n):
    sent_df = pd.read_csv(sent_csv)
    df = pd.read_csv(new_list)
    email_list = list(sent_df["email"])

    for i in range(n):
        to_send = df[~df["email"].isin(email_list)]
        details = get_random_details(to_send)
        template = get_random_template(templates)
        body, subject = template(details)
        email_list.append(details["email"])
        try:
            print(details["email"])
            send_email(subject, body, sender, details["email"], password)
        except:
            print(details["email"])
            print("Unable to send to", details)
    new_sent_df = pd.DataFrame()
    new_sent_df["email"] = email_list
    new_sent_df.to_csv(sent_csv)
```
